### Árvore de Decisão ###

```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importamos o dataset iris, que contém os dados coletados de 150 amostras de flores da mesma família
iris = load_iris()
```

```python
print(iris.DESCR)
```

```python
# Abaixo podemos visualizar os dados da chave 'data' do conjunto iris
print(iris.keys())
```

```python
# Formatei o conjunto em um dataframe pandas para facilitar a visualização
df_iris = pd.DataFrame(iris['data'], columns=iris['feature_names'])
display(df_iris)
```

```python
# Aqui definimos que as variáveis independentes serão a 3ª e 4ª coluna do dataframe
x = iris.data[:, 2:]

y = iris['target']
```

```python
tree_clf = DecisionTreeClassifier()
tree_clf.fit(x, y)
```

```python
fig, ax = plt.subplots(figsize = (15, 8))
tree.plot_tree(tree_clf)
```
